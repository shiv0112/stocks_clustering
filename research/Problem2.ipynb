#### Data Standardization

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
import seaborn as sns
from sklearn.decomposition import PCA

sns.set_style('whitegrid')
pca = PCA().fit(X_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca = PCA(n_components=10)
pca.fit(X_std)
print('explained variance :')
print('--------------------------------------------------------------------')
print(pca.explained_variance_)
print('--------------------------------------------------------------------')
print('PCA Components : ')
print('--------------------------------------------------------------------')
print(pca.components_)
print('--------------------------------------------------------------------')
X_transformed = pca.transform(X_std)
print('Transformed Feature values first five rows :')
print('--------------------------------------------------------------------')
print(X_transformed[:5,:])
print('--------------------------------------------------------------------')
print('Transformed Feature shape :')
print('--------------------------------------------------------------------')
print(X_transformed.shape)
print('--------------------------------------------------------------------')
print('Original Feature shape :')
print('--------------------------------------------------------------------')
print(X_std.shape)
print('--------------------------------------------------------------------')
print('Restransformed Feature shape :')
print('--------------------------------------------------------------------')
X_retransformed = pca.inverse_transform(X_transformed)
print(X_retransformed.shape)
print('--------------------------------------------------------------------')
print('Retransformed Feature values first five rows :')
print('--------------------------------------------------------------------')
print(X_retransformed[:5,:])
print('--------------------------------------------------------------------')
```

```python
X_trans = pd.DataFrame(X_transformed) 
X_trans.set_index(X.index,inplace=True)
```

```python
correl=pd.DataFrame(X_trans.T.corr())
```

### Standard deviation is the statistical measure of market volatility, measuring how widely prices are dispersed from the average price.

```python
X_trans['Std. Deviation'] = X_trans.T.std()
```

```python
X_trans
```

```python
std_dev=pd.DataFrame(X_trans['Std. Deviation'])
std_dev.set_index=X_trans.index
std_dev.head()
```

### K-Means

```python
from sklearn.cluster import KMeans

wcss=[]
for i in range(1, 21):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(std_dev)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 21), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Mean Squared Errors')
plt.figure(figsize=(10,8))
plt.show()
```

```python
import scikitplot
scikitplot.cluster.plot_elbow_curve(KMeans(),std_dev,cluster_ranges=range(1,20))
```

### K-Means with 3 clusters

```python
k_means = KMeans(n_clusters=3,random_state=42,init='k-means++')
X_trans['cluster number'] = k_means.fit_predict(std_dev)
X_trans['cluster number'].unique()
```
